### Creating a popularity based recommendation system

```python
#merging the datasets
ratings=Ratings.merge(Books,on='ISBN')
```

```python
num_ratings_df=ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
'''A data frame with 
tow columns(Book-Rating) where Book-Rating is the count of the ratings per boook.
Count() ignores the NaNs'''
```

```python
num_ratings_df.rename(columns={'Book-Rating':'Num-Ratings'},inplace=True)
```

```python
num_ratings_df
```

```python
# ...existing code...
avg_rating = ratings.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
# ...existing code...
```

```python
avg_rating
```

```python
popularity_df=num_ratings_df.merge(avg_rating,on='Book-Title')
```

```python
popularity_df=popularity_df[popularity_df['Num-Ratings']>250].sort_values('Book-Rating',ascending=False)
```

```python
popularity_df=popularity_df.merge(Books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Num-Ratings','Book-Rating']]
```

## COLLABORATIVE FILTERING BASED

```python
x=ratings.groupby('User-ID').count()['Book-Rating']>200
proper_users=x[x].index
```

```python
filtered_rating=ratings[ratings['User-ID'].isin(proper_users)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
# Filter pt to only famous books
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt = pt.loc[famous_books]  # Keep only famous books
pt.fillna(0, inplace=True)

# Now compute similarity with batch processing
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

batch_size = 100
n = pt.shape[0]

sim_matrix = np.memmap("similarity.dat", dtype="float32", mode="w+", shape=(n, n))

for i in range(0, n, batch_size):
    end = min(i + batch_size, n)
    sim = cosine_similarity(pt.iloc[i:end].values, pt.values)
    sim_matrix[i:end] = sim.astype("float32")

print(f"Similarity matrix shape: {sim_matrix.shape}")
```

```python
sim_matrix
```

```python
def recommend(book_name, top_n=5):
    #index fetch
    if book_name not in pt.index:
        print(f"'{book_name}' not found in dataset")
        return
    
    index=np.where(pt.index==book_name)[0][0]
    #Get similarities for the correct book (use index, not 0)
    similar_items=sorted(list(enumerate(sim_matrix[index])),key=lambda x:x[1],reverse=True)[1:top_n+1]
    
    print(f"Books similar to '{book_name}':")
    data=[]
    for i in similar_items:
        item=[]
        temp_df=Books[Books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        #print(temp_df.drop_duplicates('Book-Title')['Image-URL-M'])
        data.append(item)
    return data
```
